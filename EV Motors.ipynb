```python
#Renaming column names and adding their units 
df.rename(columns={'Accel':'AccelerationSec','Range':'Rangekm'},inplace=True)
```

```python
df.rename(columns={'Efficiency':'EfficiencyWhKm','FastCharge':'FastChargekmH','TopSpeed':'TopSpeedkmH'},inplace=True)
```

```python
df.head()
```

```python
# Replacing Rapid Changing Possiblle or not possible to yes or no
df['RapidCharge']=df['RapidCharge'].apply(lambda x: 'Yes' if x == 'Rapid charging possible' else 'No')
```

```python
df.head()
```

```python
#Removing unused column name "SEGMENT"
df.drop(columns=['Segment'],inplace=True)
```

```python
df.head()
```

```python
#Adding a new ID column

# Ensure the length of the DataFrame
n = len(df)

# Generate a list of IDs
id_list = [f'EV-{str(i).zfill(3)}' for i in range(1, n + 1)]

# Add the ID column to the existing DataFrame
df['ID'] = id_list
# Display the updated DataFrame
print("\nUpdated DataFrame with ID Column:")
print(df)
```

```python
df.head()
```

```python
#Reordering the cells
df=df[['ID','Brand','Model','Seats','PriceEuro','AccelerationSec','TopSpeedkmH','Rangekm','EfficiencyWhKm','FastChargekmH','RapidCharge','PowerTrain','PlugType','BodyStyle']]
```

```python
df.head()
```

```python
# Removing units from column values
df['AccelerationSec'] = df['AccelerationSec'].str.replace(' sec', '', regex=False)
df['AccelerationSec'] = pd.to_numeric(df['AccelerationSec'], errors='coerce')

df['TopSpeedkmH'] = df['TopSpeedkmH'].str.replace(' km/h', '', regex=False)
df['TopSpeedkmH'] = pd.to_numeric(df['TopSpeedkmH'], errors='coerce')

df['Rangekm'] = df['Rangekm'].str.replace(' km', '', regex=False)
df['Rangekm'] = pd.to_numeric(df['Rangekm'], errors='coerce')

df['EfficiencyWhKm'] = df['EfficiencyWhKm'].str.replace(' Wh/km', '', regex=False)
df['EfficiencyWhKm'] = pd.to_numeric(df['EfficiencyWhKm'], errors='coerce')

df['FastChargekmH'] = df['FastChargekmH'].str.replace(' km/h', '', regex=False)
df['FastChargekmH'] = pd.to_numeric(df['FastChargekmH'], errors='coerce')
```

```python
df['FastChargekmH']=df['FastChargekmH'].fillna(0).astype(int)
```
